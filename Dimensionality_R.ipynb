```python
import pandas
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error


df = pandas.read_csv('Datasets/winequality-red.csv',sep=";")
df
```

# Entrenamiento y prueba

```python


X = df.values[:,:-1]
Y = df.values[:,-1]
Xtrain, Xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.30, random_state=42)
print('X',X.shape)
print('Xtrain',Xtrain.shape)
print('Xtest',Xtest.shape)
```

# Correlacion entre variables de entrada y Quality

```python
Co = {}

for i in range (11):
    cols= Xtrain[:,i]
    corr= abs( np.corrcoef(cols,ytrain)[0,1] )
    Co[i]=corr
    print(corr)

print()
print("Correlaciones[]: ")
print(Co)
    
```

```python
sort_co = sorted(Co.items(), key=lambda kv: kv[1]) # Orden Ascendete
sort_co =sort_co[::-1]                             # Orden Descendente 
sort_co
```

# Modelo MLP 

```python
from sklearn.metrics import accuracy_score
MLP=[]
cont=1
for i in sort_co:
    MLP.append(i[0])
    model = MLPRegressor()
    model.fit(Xtrain[:,MLP],ytrain)
    ypred = model.predict(Xtest[:,MLP])
    
    
    print("MLP con: ", cont )
    print()
    print("Indices: ",MLP)
    print('Ytest',ytest[:10])
    print('Ypred',ypred[:10])
    print('MAE:',mean_absolute_error(ytest,ypred))
    print('MSE:',mean_squared_error(ytest,ypred))
    print('Y values:',min(Y),max(Y))
    print('Accuracy',accuracy_score(ytest,ypred.round()))
    print()
    cont+=1

    
```

###  Indices:  [10, 1, 9, 6, 2]   (alcohol, volatile acidity, sulphates, total sulfur dioxide, citric acid) fueron los indicadores que aportaron mayor precisión con un: Accuracy 0.5666666666666667

# Modelo Decision Tree Regressor

```python
DTR=[]
cont=1
for i in sort_co:
    DTR.append(i[0])
    model = DecisionTreeRegressor()
    model.fit(Xtrain[:,DTR],ytrain)
    ypred = model.predict(Xtest[:,DTR])
    print("DTR con: ", cont )
    print()
    print("Indices: ",DTR)
    print('MAE:',mean_absolute_error(ytest,ypred))
    print('MSE:',mean_squared_error(ytest,ypred))
    print('Y values:',min(Y),max(Y))
    print('Accuracy',accuracy_score(ytest,ypred.round()))
    print()
    cont+=1
```

### Indices:  [10, 1, 9, 6, 2, 7, 4, 0, 5]  (alcohol, volatile acidity, sulphates, total sulfur dioxide, citric acid, density, chlorides, fixed acidity, free sulfur dioxide) fueron los indicadores con los que la precisión suele ser mas efectiva con un: Accuracy 0.6229166666666667
