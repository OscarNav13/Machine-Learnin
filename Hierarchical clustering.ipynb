```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import pandas  
import numpy as np
from scipy.cluster import hierarchy

df = pandas.read_csv('movies_evaluations.csv')
df
```

```python
data=df.to_numpy()
print(data.shape)
X=data[:,1:len(data)+1]
Mnames=data[:,0]
print(X.shape)
print(Mnames)
```

```python

hierarchy.set_link_color_palette(['plum','lightseagreen', 'silver'])
Z = hierarchy.linkage(X)
#print(Z)
dn = hierarchy.dendrogram(Z,labels=Mnames)
plt.show()
```
