```python
import pandas                   # Leer el archivo csv
import numpy as np              # Operaciones con vectores y matrices
import matplotlib.pyplot as plt # Gráficas
```

```python
df = pandas.read_csv('adult.csv',sep=",")
df
```

```python
var = 2
name = df.columns[var]
values = df.values[:,var]
print(name)
print(values)
```

```python
# Frecuencias
vunique, vcounts = np.unique( values, return_counts=True )
print(vunique)
print(vcounts)
```

```python
# Entropy
def entropy(X):
    # H = - Sum ( P(x) log( P(x)))
    h = 0
    n = len(X)
    vunique, vcounts = np.unique( X, return_counts=True )
    for u, c in zip( vunique, vcounts):
        prob = c / n
        h += prob * np.log2(prob)
    return -h , - np.log2( 1/len(vunique) )

h = entropy(values)
print(h)
```

```python
for var in range(len(df.columns)):
    name = df.columns[var]
    values = df.values[:,var]
    h, hM = entropy(values)
    print(name,h,hM)
```

```python
def mutual_information(X, Y):
    x_ = df.columns[X] # Header
    y_ = df.columns[Y] # Header
    
    uniqueX = np.unique(df.values[:,X], return_counts = False) # Types
    uniqueY = np.unique(df.values[:,Y], return_counts = False) # Types
    
    total = len(df.values[:,X]) # number of rows
    
    n = len(uniqueX) # number of distinct types
    m = len(uniqueY)    

    Px, Py = 0,0
    Pxy = 0
    mi = 0
    for i in range(0,n):
        Px = len(df[df[x_]==uniqueX[i]])/total
        #print(Px)
        for j in range(0,m):
            Py = len(df[df[y_]==uniqueY[j]])/total
            
            cant = len(df[(df[x_]==uniqueX[i]) & (df[y_]==uniqueY[j])]) # Intersection between x and y
            Pxy = cant/total
            #print(Pxy)
            if (Pxy/(Px*Py)<=0):
                mi += 0
            else:
                mi += (Pxy * np.log2(Pxy /(Px*Py)))
    
    return mi

for var1 in range(len(df.columns)):
    for var2 in range(var1+1,len(df.columns)):

        name1 = df.columns[var1]
        values1 = df.values[:,var1]
        h1, hM1 = entropy(values1)
        
        name2 = df.columns[var2]
        values2 = df.values[:,var2]
        h2, hM2 = entropy(values2)

        
        im = mutual_information(var1,var2)
        
        print(name1,h1,name2,h2,im)
```
