```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import cv2
from scipy.cluster import hierarchy
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

```python
img = cv2.imread('img.jpg')

plt.figure()
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

# Vectorizar la imagen
nrows,ncols,nch = img.shape
X = np.reshape(img,(nrows*ncols,3))
print(X.shape)
X
```

# K means metod

```python

model = KMeans(n_clusters=3)
model.fit(X)

# Metemos los datos al modelo y sacamos la prediccion
group = model.predict(X)
print(model.cluster_centers_)
```

## K3

```python
# Creamos un arreglo y reagrupamos los datos para hacer un proceso inverso, para asi mostrar la imagen predecida
imgK=[]
for i in group:
    if  (i==0):
        imgK.append(model.cluster_centers_[0])
    elif (i==1):
        imgK.append(model.cluster_centers_[1])
    elif (i==2):
        imgK.append(model.cluster_centers_[2])
imgC = np.array(imgC,np.uint8)
```

```python
PredImg= np.reshape(imgC,(nrows,ncols,3))
plt.figure()
plt.imshow(cv2.cvtColor(PredImg, cv2.COLOR_BGR2RGB))
plt.show()
```
