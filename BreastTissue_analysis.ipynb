```python
var = 2
print( df.columns[var])
columna = df.values[:,var]
print('Minimo',np.min(columna))
print('Maximo',np.max(columna))
print('Promedio',np.mean(columna))
print('Varianza',np.var(columna))
print('Desv estándar',np.std(columna))
print('Mediana',np.median(columna))
print('Q0 Cuartil 0',np.quantile(columna,0.0))
print('Q1 Cuartil 1',np.quantile(columna,0.25))
print('Q2 Cuartil 2',np.quantile(columna,0.5))
print('Q3 Cuartil 3',np.quantile(columna,0.75))
print('Q4 Cuartil 4',np.quantile(columna,1.0))
q3 = np.quantile(columna,0.75)
q1 = np.quantile(columna,0.25)
IQR = q3-q1
print('Rango intercuartil IQR Q3-Q1', IQR)
print('Atipico  < Q1 - 1.5*IQR', q1 - 1.5*IQR)
print('Atipico  > Q3 + 1.5*IQR', q3 + 1.5*IQR)
```

```python
import matplotlib.pyplot as plt
```

```python
var = 2
print( df.columns[var])
columna = df.values[:,var]

plt.figure()
plt.hist(columna,bins=20)
plt.show()

plt.figure()
plt.boxplot(columna)
plt.show()
```

```python
for var in range(2,11):
    columna = df.values[:,var]
    plt.figure()
    plt.suptitle(df.columns[var])
    plt.subplot(1,2,1)
    plt.hist(columna,bins=80)
    plt.subplot(1,2,2)
    plt.boxplot(columna)
    plt.show()
```

```python
var1 = 2
var2 = 3
columna1 = df.values[:,var1]
columna2 = df.values[:,var2]



print('Covarianza:',np.cov(columna1.astype(float),columna2.astype(float)),np.cov(columna1.astype(float),columna2.astype(float)))
print('Correlación:',np.corrcoef(columna1.astype(float),columna2.astype(float)),np.corrcoef(columna1.astype(float),columna2.astype(float)))

corr = np.corrcoef(columna1.astype(float),columna2.astype(float))
plt.figure()
plt.title('Corr:'+str(corr))
plt.xlabel(df.columns[var1])
plt.ylabel(df.columns[var2])
plt.plot(columna1,columna2,'.')
plt.show()
```

```python
for var1 in range(2,11):
    for var2 in range(var1+1,11):
        columna1 = df.values[:,var1]
        columna2 = df.values[:,var2]
        
        

        corr = np.corrcoef(columna1.astype(float),columna2.astype(float))
        plt.figure()
        plt.title('Corr:'+str(corr))
        plt.xlabel(df.columns[var1])
        plt.ylabel(df.columns[var2])
        plt.plot(columna1.astype(float),columna2.astype(float),'.')
        plt.show()
```
