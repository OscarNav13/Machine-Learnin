```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import cv2
from scipy.cluster import hierarchy
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

```python
img = cv2.imread('power.jpg')
plt.figure()
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
# Vectorizar la imagen
nrows,ncols,nch = img.shape
X = np.reshape(img,(nrows*ncols,3))
```

# K means

```python
model = KMeans(n_clusters=3)
model.fit(X)
group = model.predict(X)
print(model.cluster_centers_)
```

```python
imgC=[]
for i in group:
    if (i==0):
        imgC.append(model.cluster_centers_[0])
    elif (i==1):
        imgC.append(model.cluster_centers_[1])
    elif (i==2):
        imgC.append(model.cluster_centers_[2])
```

```python
imgC = np.array(imgC,np.uint8)
print(imgC.shape)
```

## k=3

```python
newImage= np.reshape(imgC,(nrows,ncols,3))
plt.figure()
plt.imshow(cv2.cvtColor(newImage, cv2.COLOR_BGR2RGB))
plt.show()
```

## k=10

```python
model = KMeans(n_clusters=10)
model.fit(X)
group = model.predict(X)
imgC=[]
for i in group:
    if (i==0):
        imgC.append(model.cluster_centers_[0])
    elif (i==1):
        imgC.append(model.cluster_centers_[1])
    elif (i==2):
        imgC.append(model.cluster_centers_[2])
    elif (i==3):
        imgC.append(model.cluster_centers_[3])
    elif (i==4):
        imgC.append(model.cluster_centers_[4])
    elif (i==5):
        imgC.append(model.cluster_centers_[5])
    elif (i==6):
        imgC.append(model.cluster_centers_[6])
    elif (i==7):
        imgC.append(model.cluster_centers_[7])
    elif (i==8):
        imgC.append(model.cluster_centers_[8])
    elif (i==9):
        imgC.append(model.cluster_centers_[9])
  
        
imgC = np.array(imgC,np.uint8)
newImage= np.reshape(imgC,(nrows,ncols,3))
plt.figure()
plt.imshow(cv2.cvtColor(newImage, cv2.COLOR_BGR2RGB))
plt.show()
```

# Gaussian Mixture

## k=3

```python
model = GaussianMixture(n_components=3)
model.fit(X)

group=model.predict(X)

imgC=[]
for i in group:
    if (i==0):
        imgC.append(model.means_[0])
    elif (i==1):
        imgC.append(model.means_[1])
    elif (i==2):
        imgC.append(model.means_[2])

imgC = np.array(imgC,np.uint8)
newImage= np.reshape(imgC,(nrows,ncols,3))
plt.figure()
plt.imshow(cv2.cvtColor(newImage, cv2.COLOR_BGR2RGB))
plt.show()
```
